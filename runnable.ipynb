```python
from abc import ABC, abstractmethod

# base runnable class
class BaseUnit(ABC):

  @abstractmethod
  def run(data):
    pass
```

```python
import random

class DummyModel(BaseUnit):

  def __init__(self):
    print('Model initialized')

  def run(self, text):

    outputs = [
        'Delhi is the capital of India',
        'IPL is a cricket league',
        'AI stands for Artificial Intelligence'
    ]

    return {'result': random.choice(outputs)}


  def generate(self, text):

    outputs = [
        'Delhi is the capital of India',
        'IPL is a cricket league',
        'AI stands for Artificial Intelligence'
    ]

    return {'result': random.choice(outputs)}
```

```python
class DummyPrompt(BaseUnit):

  def __init__(self, pattern, variables):
    self.pattern = pattern
    self.variables = variables

  def run(self, values):
    return self.pattern.format(**values)

  def build(self, values):
    return self.pattern.format(**values)
```

```python
class TextParser(BaseUnit):

  def __init__(self):
    pass

  def run(self, data):
    return data['result']
```

```python
class Pipeline(BaseUnit):

  def __init__(self, steps):
    self.steps = steps

  def run(self, data):

    for step in self.steps:
      data = step.run(data)

    return data
```

```python
prompt_obj = DummyPrompt(
    pattern='Write a {length} poem about {topic}',
    variables=['length', 'topic']
)
```

```python
model = DummyModel()
```

```python
output_parser = TextParser()
```

```python
pipeline = Pipeline([prompt_obj, model, output_parser])
```

```python
pipeline.run({'length':'long', 'topic':'india'})
```

```python
prompt_a = DummyPrompt(
    pattern='Write a joke about {topic}',
    variables=['topic']
)
```

```python
prompt_b = DummyPrompt(
    pattern='Explain the following joke {result}',
    variables=['result']
)
```

```python
model = DummyModel()
```

```python
output_parser = TextParser()
```

```python
stage1 = Pipeline([prompt_a, model])
```

```python
stage2 = Pipeline([prompt_b, model, output_parser])
```

```python
complete_pipeline = Pipeline([stage1, stage2])
```

```python
complete_pipeline.run({'topic':'cricket'})
```
